```python
import pandas as pd 
import re
```

```python
amazon = pd.read_csv(r"C:\Users\Divya\skills\machineL\deep_learning\dataset\items\amazon_review.csv")
flipkart = pd.read_csv(r"C:\Users\Divya\skills\machineL\deep_learning\dataset\items\all_flipkart_products.csv")
imdb = pd.read_csv(r"C:\Users\Divya\skills\machineL\deep_learning\dataset\items\all_imdb_reviews.csv")
zomato = pd.read_csv(r"C:\Users\Divya\skills\machineL\deep_learning\dataset\items\all_zomato_reviews.csv")
```

```python
amazon.columns = amazon.columns.str.strip().str.lower()
flipkart.columns = flipkart.columns.str.strip().str.lower()
imdb.columns = imdb.columns.str.strip().str.lower()
zomato.columns = zomato.columns.str.strip().str.lower()
```

```python
amazon = amazon[["review", "label", "source"]]
flipkart = flipkart[["review", "label", "source"]]
imdb = imdb[["review", "label", "source"]]
zomato = zomato[["review", "label", "source"]]
```

```python
amazon["source"] = "amazon"
flipkart["source"] = "flipkart"
imdb["source"] = "imdb"
zomato["source"] = "zomato"
```

```python
combined = pd.concat([amazon, flipkart, imdb, zomato], ignore_index=True)
```

```python
combined = combined.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
combined.head()
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text) 
    return text.strip()

combined["review"] = combined["review"].apply(clean_text)
```

```python
combined.to_csv("dataset/cleaned_data.csv", index=False)
```
