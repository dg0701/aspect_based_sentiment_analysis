```python
import nltk
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

rows = []

for _, row in df.iterrows():
    
    review = row["review"]
    aspect = row["aspect"]

    words = review.lower().split()

    if aspect.lower() in words:
        idx = words.index(aspect.lower())

        start = max(0, idx-5)
        end = min(len(words), idx+5)

        context = " ".join(words[start:end])
    else:
        context = review

    score = sia.polarity_scores(context)["compound"]

    if score > 0.05:
        sentiment = "positive"
    elif score < -0.05:
        sentiment = "negative"
    else:
        sentiment = "neutral"

    rows.append({
        "review": review,
        "aspect": aspect,
        "sentiment": sentiment
    })

new_df = pd.DataFrame(rows)
new_df.head
```

```python
new_df.to_csv("aspect_sentiment_dataset.csv", index=False)
```
