```python
import pandas as pd

df = pd.read_pickle("final_absa_dataset.pkl")

print(df.head())
print(len(df))
```

```python
df.head
```

```python
print(df["aspect_sentiment"].value_counts())
```

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(df,test_size=0.2,stratify=df["aspect_sentiment"],random_state=40)
```

```python
val_df, test_df = train_test_split(temp_df, test_size=0.5,random_state=40)

print(len(train_df), len(val_df), len(test_df))
```

```python
import os
os.makedirs("dataset/train", exist_ok=True)
os.makedirs("dataset/val", exist_ok=True)
os.makedirs("dataset/test", exist_ok=True)
```

```python
train_df.to_pickle("dataset/train/train.pkl")
val_df.to_pickle("dataset/val/val.pkl")
test_df.to_pickle("dataset/test/test.pkl")
```
