```python
import pandas as pd 
import numpy as np
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv(r"C:\Users\Divya\skills\machineL\deep_learning\project\aspect_based_sentiment_analysis\aspect_sentiment_dataset.csv")
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
def convert_sentiment(text, label):
    if label != 'neutral':
        return label
    score = sia.polarity_scores(text)['compound']
    if score >= 0:
        return 'positive'
    else:
        return 'negative'

df['sentiment'] = df.apply(lambda x: convert_sentiment(x['review'], x['sentiment']), axis=1)
```
