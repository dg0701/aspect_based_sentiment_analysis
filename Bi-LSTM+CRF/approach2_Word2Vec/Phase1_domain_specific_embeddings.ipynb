```python
from gensim.models import Word2Vec
import pandas as pd
from tqdm import tqdm
```

```python
data = pd.read_csv("cleaned_data.csv")
data.shape
```

```python
#Extraction of reviews:
reviews = data['review']
reviews = list(reviews)
reviews.__len__()
```

```python
cleaned_reviews = []
for r in reviews:
    if isinstance(r, str) and len(r.strip()) > 0:
        cleaned_reviews.append(r.lower().split()) 
    elif isinstance(r, list):
        cleaned_reviews.append(r)
```

```python
model_w2v = Word2Vec(vector_size = 300,
                    window=5,
                    min_count=8,
                    workers=4,
                    sg=1)
print("Building Vocabulary...")
model_w2v.build_vocab(tqdm(cleaned_reviews, desc="Vocab Progress"))
```

```python
print("Training Model...")
model_w2v.train(
    tqdm(cleaned_reviews, desc="Training Progress"), 
    total_examples=len(cleaned_reviews), 
    epochs=10
)

model_w2v.save("hinglish_absa_w2v.model")
```

```python
print("Similar to 'delivery':", model_w2v.wv.most_similar("delivery"))

try:
    print("Similar to 'buri':", model_w2v.wv.most_similar("buri"))
except KeyError:
    print("'achha' not found in vocabulary.")
```
